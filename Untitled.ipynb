## 1. Patient
- infection_order: 第幾號傳染者，1號代表來自武漢，N號代表是被N-1號傳染
- contact_number:確診者接觸人數
- state: 
 - isolated:在醫院中隔離
 - released:康復
 - deceased:死亡
- age:年齡
- disease: TRUE: underlying disease / FALSE: no disease




```python
display(patientInfo.head(20).append(patientInfo.tail(5)))
```

```python
region = pd.DataFrame()
count = patientInfo.groupby('province').count()

stat = patientInfo.contact_number.describe()
print(stat)
region['region'] = count.index
region['case'] = count.patient_id.values
region = region.sort_values('case', ascending=False)
region
```

```python
#將100以下合併
new_row = {'region':'Others', 'case':region.iloc[7:,1].sum()}
region = region.append(new_row, ignore_index=True).sort_values('case', ascending=False)[0:8]
region
```

```python
#印出各地區確診人數
fig, ax = plt.subplots(figsize=(11, 11))
plt.title(f'Cases in each Province', fontsize=15)
plt.pie(region.case
        , labels=region.region
        , autopct='%.2f%%'
        , startangle=90
        , counterclock=False)
plt.show()
```
